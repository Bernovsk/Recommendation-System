### 1.1 _Problem_

_Currently applications that are created to improve or facilitate people's lives are launched at all times. A possible idea to add to this group of applications, would be the development of an ML model that provides a certain person  the best place for them to move  in a certain neighborhood. Supose that this person wants to move to live in Ipanema in Rio de Janeiro, Brazil, she considers essential that near her new home should have restaurants, cafeterias and markets, based on these preferences the system would inform her in what part of this neighborhood would be more compatible with their preferences. This project aims to divide certain neighborhoods into 5 areas and then analyze what is most common in the locations in each area, followed by a recommendation of the most compatible location based on the person's preferences._

In this case, I will use a composite model, where I will first try to apply a K-MEAN model to group the locations that are similar and then I will create a recommendation system based on the greater compatibility of the clusters with the person preferences, then I will provide the 5 most compatible locations for the person._

### Pyproj

 - _pyproj convert latitudes and longitudes into cartesian coordinates (lonlat_to_xy(), meters), as the vice versa case (xy_to_lonlat())._
 - _The third function retrieve the geometric distance in the cartesian plane._


```python
from functools import partial
from pyproj import Proj
from pyproj import Transformer
from pyproj import CRS

def lonlat_to_xy(lon, lat):
    crs_4326 = CRS("WGS84")
    crs_proj = CRS("EPSG:5641")
    transformer = Transformer.from_crs(crs_4326, crs_proj, always_xy = True)
    xy = transformer.transform(lon, lat)
    
    return xy[0], xy[1]

def xy_to_lonlat(x, y):
    crs_4326 = CRS("WGS84")
    crs_proj = CRS("EPSG:5641")
    transformer = Transformer.from_crs(crs_proj, crs_4326, always_xy = True)
    lonlat = transformer.transform(x, y)
    

    return lonlat[0], lonlat[1]


def calc_xy_distance(x1, y1, x2, y2):
    dx = x2 - x1
    dy = y2 - y1
    return math.sqrt(dx*dx + dy*dy)

print('Coordinate transformation check')
print('-------------------------------')
print('Ipanema center longitude={}, latitude={}'.format(Ipanema[1], Ipanema[0]))

ipanema_center_x, ipanema_center_y = lonlat_to_xy(Ipanema[1],Ipanema[0])
print('Ipanema center UTM X={}, Y={}'.format(ipanema_center_x, ipanema_center_y))

longitude, latitude = xy_to_lonlat(ipanema_center_x, ipanema_center_y)
print('Ipanema center longitude={}, latitude={}'.format(longitude, latitude))
```

### _Addresses grid around Ipanema Neighborhood_
- _It is important to note that the circles are interpolated, which will allow you to cover as much of the region as possible without leaving any uncovered areas._

```python
def get_sw(coordinates, x_offset_min, y_offset_min): #X min Y min
    x_center, y_center = lonlat_to_xy(coordinates[1], coordinates[0])
    x_sw = x_center - x_offset_min
    y_sw = y_center - y_offset_min
    
    lon_sw, lat_sw = xy_to_lonlat(x_sw, y_sw)
    
    return x_sw, y_sw #, lon_sw, lat_sw

def get_se(coordinates, x_offset_max, y_offset_min): #X max Y min
    x_center, y_center = lonlat_to_xy(coordinates[1], coordinates[0])
    x_se = x_center + x_offset_max
    y_se = y_center - y_offset_min
    
    lon_se, lat_se = xy_to_lonlat(x_se, y_se)
        
    
    return x_se, y_se #, lon_se, lat_se


def get_ne(coordinates, x_offset_max, y_offset_max): #X max Y max
    x_center, y_center = lonlat_to_xy(coordinates[1], coordinates[0])
    x_ne = x_center + x_offset_max
    y_ne = y_center + y_offset_max
    
    lon_ne, lat_ne = xy_to_lonlat(x_ne, y_ne)
        

    return x_ne, y_ne #, lon_ne, lat_ne



def get_nw(coordinates, x_offset_min, y_offset_max): #X min Y max
    x_center, y_center = lonlat_to_xy(coordinates[1], coordinates[0])
    x_nw = x_center - x_offset_min
    y_nw = y_center + y_offset_max
    
    lon_nw, lat_nw = xy_to_lonlat(x_nw, y_nw)
        

    return  x_nw, y_nw #, lon_nw, lat_nw    
```

```python
def grid_array(Init_coord, x_min, y_min, x_max, y_max, x_steps, y_steps, radius):
    latitude = []
    longitude = []
    
    distances_from_center = []
    xs = []
    ys = []
    
    coord_center_x, coord_center_y = lonlat_to_xy(Init_coord[1],
                                                  Init_coord[0]) # Neighborhood center in Cartesian coordinates
    #The, i choose to interate over the SW direction to the NE
    x_sw, y_sw = get_sw(Init_coord, x_min, y_min) 
    
    x_ne, y_ne = get_ne(Init_coord, x_max, y_max)
    

    radius_x =(x_ne - x_sw)/(2*x_steps)
    
    radius_y =(y_ne - y_sw)/(2*y_steps)
    
    #lon_pos, lat_pos = array_calculus(x_sw, radius, radius, n_x, n_y)
    
    for i in range(x_steps):
        x_pos = x_sw + i * 2* radius
        for j in range(y_steps):
            y_pos = y_sw + j * 2 *  radius
            lon_pos, lat_pos = xy_to_lonlat(x_pos, y_pos)
            
            distance_from_center = calc_xy_distance(coord_center_x, coord_center_y, x_pos, y_pos)
            longitude.append(lon_pos)
            latitude.append(lat_pos)
            distances_from_center.append(distance_from_center)
            xs.append(x_pos)
            ys.append(y_pos)
            
    #Location_dict = {'Latitudes': latitude, 'Longitudes': longitude,
    #                 'Distances': distances_from_center, 'xs':xs, 'ys':ys}      
    lat_f = []
    lon_f = []
    xs_f = []
    ys_f = []
    distances_from_center_list_f = []  
    google_api_key = "AIz####################5A"
    for i in range(len(latitude)):
        Coordinates = coordinates_filter(latitude[i], longitude[i], google_api_key)
        lati = Coordinates[0]
        longi = Coordinates[1]
        bool_test = lati is not None
        if bool_test:
            lat_f.append(lati)
            lon_f.append(longi)
            xs_f.append(xs[i])
            ys_f.append(ys[i])
            distances_from_center_list_f.append(distances_from_center[i]) 
    Location_dict = {'Latitudes': lat_f, 'Longitudes': lon_f,
                     'Distances': distances_from_center_list_f, 'xs':xs_f, 'ys':ys_f}  
            
    return Location_dict
```

```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=350):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        #print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

## Methodology <a name="item7"></a>


**The methodology will contain two models:**

 - _K-MEANS machine learning model, because we are trying to model categorical data to get an collaborative recommendation system._ 
 - _The second model is a based content recommendation system._
 
### Based Content Recommendation System
 
 - _I had simulate randomly 50 customers profile._
 - _Suppose that our customers had provide their prefereences in level of importance to them. Every level of importance will provide some weight ($W_i$ in the E. 1 bellow) to their choose. Based on this our customer profile is represented by the following matrix:_
     
$$ Customer_{profile}= \begin{pmatrix} W_1.Venue_1 & W_2.Venue_2 & ... & W_n.Venue_n \end{pmatrix} (E. 1)$$

  - _If we take the product between $Customer_{profile}$ and the correspondent matrix of address density Venues ($M\rho_{Venues}$, where each row represents the correspondent Address density of venues values), we will get:_.

$$ M\rho_{Venues} X {Customer_{profile}}^T =  \begin{pmatrix} \rho_{Venue_1} & \rho_{Venue_2} & ... & \rho_{Venue_n} \\ \rho_{Venue_1} & \rho_{Venue_2} & ... & \rho_{Venue_n} \\. & . & . & .\\. & . & .& .\\ . & . & .& . \\ \rho_{Venue_1} & \rho_{Venue_2} & ... & \rho_{Venue_n}\end{pmatrix} X \begin{pmatrix} W_1.Venue_1 \\ W_2.Venue_2 \\ ... \\ W_n.Venue_n \end{pmatrix} = \begin{pmatrix} Score_{Address_1} \\ Score_{Address_2} \\ Score_{Address_3}\\. \\. \\. \\ Score_{Address_n}  \end{pmatrix}$$


 - _Where the ${Score}_{Address_i}$ is the score value for each address:_
 
 $$ {Score}_{Address_i} = \rho_{Venue_1} * W_1.Venue_1 + \rho_{Venue_2} * W_2.Venue_2 + \rho_{Venue_3} * W_3.Venue_3 + ... + \rho_{Venue_n} * W_n.Venue_n  $$
 
 
 - _The highest  $ {Score}_ {Address_i} $  comprises the most compatible to the customer profile._

###  K means <a name="item8"></a>

- _It is an Unsupervised Learning technique, where it has a series of blobs in my parameter space, and it adjusts to each iteration based on the distance from each data point._
- _First, we have to turn our data in categorical type, since this unsupervised model try to create label in groups of similar data (addresses)_.

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

 - _Generate a list of values ranged from 0.5 to 0.0001. This list will be the weights of the choices of the customer, and will give randomly to each value in the Customer_preferences. In other words, this is a way to think that the most priority things that the customer need is equal to 0.5 and the least important is 0.0156._
 - _Zero is to increase the random when I multiply between the customer profile and the neighborhood hot encoding dataframe._

```python
n = 0.5
List = [0, 0.5]
for i in range(5):
    n = 0.5*n
    if np.sum(List) >= 1:
        break
    else:
        List.append(n)
```

```python
for index, names in enumerate(Venues_category):

    for i in range(number_customer):
        Customer_preferences[names].loc[i:i] = Customer_preferences[names] * random.choice(List)
```

#### _Customers preferences dataset_

```python
Customer_preferences.info()
Customer_preferences
```

```python
Address_line, Addess_column = Ipanema_grouped.shape
```

```python
Ipanema_grouped.describe()
```

- _The next step, was to multiply the weighted values of the customer  by the density of the correspondent venue in each address._

```python
Customers_compatibility_points = {}

for i in range(Line): #This Loop will perfomed in each customer row
    Score = {}
    Address_score = []
    for j in range(Address_line): #This loop will parse each address row in Ipanema Grouped
        
        # Val list is the score list from each Neighborhood
        Addr = Ipanema_grouped['Neighborhood'].loc[j:j].item() #Address
        Val = []
        
        for names in Venues_category: #This loop will parse each Venue Category
            Cust_preferences = Customer_preferences[names].loc[i:i].item() #Customer preferences 
            Place_density = Ipanema_grouped[names].loc[j:j].item() # Venues density in each Address
            
            Value = Cust_preferences * Place_density #Dot product (matrix product) of the customer and place density
            Val.append(Value)
            
        Sum_scores = np.sum(Val)  #final sum of our matrix product
        
        Score.update({Addr:Sum_scores})
        
    Customers_compatibility_points[i] = Score
```

```python
Customer_compatibility = pd.DataFrame(Customers_compatibility_points)
```

```python
Final_score = Customer_compatibility
```

```python
for i in range(number_customer):
    Customer_ = Final_score[i].sort_values(ascending = False).to_frame()
    Customer_.reset_index(inplace = True)
    Customer_.rename(columns={'index':'Address', i:'Costumer '+ str(i+1)}, inplace = True)
    #Customer_.rename(columns={str(i):'Costumer '+ str(i)}, inplace = True)
    print(Customer_.head())
    print('_________________________________________________', '\n')
```
